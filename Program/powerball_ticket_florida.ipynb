# Generate the Target variable: 
### tickets_sold = winners*38.32 
### since expected winners = tickets sold * probability of winning (1/38.32) 

```python
df["tickets_sold"] = df["winners"]*38.32
print(df["tickets_sold"].head()) 
```

```python
print(df.describe())
```

```python
# histogram of tickets_sold 
plt.figure(figsize=(8,5))
plt.hist(df["tickets_sold"],bins=20)
plt.xlabel("PowerBall Tickets Sold")
plt.ylabel("Frequency")
plt.title("Histogram of Florida PowerBall Tickets Sold")
plt.show()
```

```python
# histogram for tickets sold by period
fig, axes = plt.subplots(1,2,figsize=(12,5),sharex=True,sharey=True)

# Pre_covid
axes[0].hist(df.loc[df["covid"]==0,"tickets_sold"], bins=20)
axes[0].set_title("Pre-Covid")
axes[0].set_xlabel("PowerBall Tickets Sold")
axes[0].set_ylabel("Frequency")

# Covid
axes[1].hist(df.loc[df["covid"]==1,"tickets_sold"], bins=20)
axes[1].set_title("Covid")
axes[1].set_xlabel("PowerBall Tickets Sold")
axes[1].set_ylabel("Frequency")

plt.suptitle("Histogram of PowerBall Tickets Sold by Period")
plt.tight_layout()
plt.show()
```
